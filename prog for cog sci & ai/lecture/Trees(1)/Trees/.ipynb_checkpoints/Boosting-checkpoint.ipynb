## Boosting
 
* A general approach that can be applied to many statistical learning methods for regression or classification. 
* Similar to bagging except that the trees are grown sequentially
* Each tree is grown using error information from previously grown trees.
* The samples that the previos tree performed the worst on are given extra weight
    - For regression: the samples that had the highest error
    - For classificationL the sample taht were misclassified 
 

### Weak learner
 
* A learning algorithm that performs slightly better than chance
        - e.g. a tree with one split, called a stump
* Boosting combines a series of weak learners to form one strong learner 
* Each weak learner improves on prediction on part of the space

 
 
### Boosting algorithm for regression trees
 
* Fit each model to the residuals of the previous model

#### Algorithm
 
  1. Set $\hat{F}(x) = 0 \text{ and }r_i = y_i$  for all i in the training set. 
  
  2. For m = 1,2,...,M, (M is number of terms to add to model) repeat:  
  2.1 Fit a regression tree $\hat{f}^m$ with d splits  to r (the residuals)  
  2.2 Update $\hat{F}$ by adding in a shrunken version of the new tree: $\hat{F}(x)\leftarrow{\hat{F} + \lambda{\hat{f}^m(x)}}$  ($\lambda$ is the shrinkage parameter)
  2.3 Update the residuals: $r_i\leftarrow{r_i - \lambda{\hat{f}^m(x_i)}}$ 
  3. Output the boosted model:
$$\hat{F}(x) = \sum_{i=1}^N\lambda{\hat{f}^m(x)}$$
 

### Slow learning 

* Given the current model: Fit a tree using the current residuals, rather than Y as the response. 
* Add this new decision tree into the fitted function to update the residuals. 
    - The parameter d determines the size of the tree (d=1 is a stump)
* Fitting small trees to the residuals slowly improves $\hat{f}$ in areas where it does not perform well.
* The shrinkage parameter $\lambda$ slows the process down even further.
* In general, statistical learning approaches that learn slowly tend to perform well.

### Boosting Tuning Parameters
 
* M: the number of trees
    - Can overfit (unlike bagging or random forests) but usually it overfits slowly
    - Use cross-validation to select
* $\lambda$: Shrinkage parameters controls rate of learning
    - Typically .01 or .001
* d: number of splits
    - Controls complexity of the ensemble
    - d = 1 works well

#### Boosting in Python

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor

df = pd.read_csv('cadata.csv')
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values

# Make Validation Set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1234)

# Fit model to traing data
m = GradientBoostingRegressor(n_estimators = 500, max_depth =6, learning_rate = 0.1)
m.fit(X_train, y_train)

# Predict the test data
y_pred = m.predict(X_test)

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print("R2: %.4f" % r2)
```

```python
import matplotlib.pyplot as plt

# compute test set deviance
test_score = np.zeros((500,), dtype=np.float64)

for i, y_pred in enumerate(m.staged_predict(X_test)):
    test_score[i] = m.loss_(y_test, y_pred)

plt.figure(figsize=(12, 6))
plt.subplot(1, 1, 1)
plt.title('Deviance')
plt.plot(np.arange(500) + 1, m.train_score_, 'b-',label='Training Set Deviance')
plt.plot(np.arange(500) + 1, test_score, 'r-',label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')
plt.show()
```

```python
feature_importance = m.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
feat_names = df.columns.tolist()[1:]
sorted_names = [feat_names[i] for i in sorted_idx]
pos = np.arange(sorted_idx.shape[0]) + .5
plt.figure(figsize=(12, 6))
plt.subplot(1, 1, 1)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos,sorted_names)
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

#### AdaBoost in Python

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split


df = pd.read_csv("Default.csv")

X = df.iloc[:,1:].values
y = df.iloc[:,0].values


labelencoder = LabelEncoder()
X[:, 0] = labelencoder.fit_transform(X[:, 0])
y = labelencoder.fit_transform(y)

# Make Validation Set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1234)

# Fit model to traing data
m = AdaBoostClassifier(n_estimators = 500,  learning_rate = .1,random_state = 1234)
m.fit(X_train, y_train)

# Predict the test data
y_pred = m.predict(X_test)

print("Feature Importance: ",m.feature_importances_)
print("Confusion Matrix")
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy = np.trace(cm)/np.sum(cm)
print("Accuracy is: ", accuracy)
print("Mean accuracy from score method: ",m.score(X,y))
```
