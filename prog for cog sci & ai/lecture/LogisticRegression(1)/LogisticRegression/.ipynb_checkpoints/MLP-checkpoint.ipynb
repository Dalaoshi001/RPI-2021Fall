```python
from matplotlib.colors import ListedColormap
def classification_plot(X_data,y_data,model,ax):
    # Make grid: min to max of 1st column, min to max of 2nd column in small increments
    X1,X2 = np.meshgrid(np.arange(X_data[:, 0].min() - 1, X_data[:, 0].max() + 1, step = 0.01),
                     np.arange(X_data[:, 1].min() - 1, X_data[:, 1].max() + 1, step = 0.01))
    # Flatten X1 and X2, create an array and transpose to a 2-column array (one for each feature)
    v = np.array([X1.ravel(), X2.ravel()]).T
    #  Using fitted model, predict the points in v and reshape to a 2-dimensional array
    print(v.shape,X1.shape,X2.shape)
    print(model.predict(v).reshape(X1.shape).shape)
    
    ax.contourf(X1, X2, model.predict(v).reshape(X1.shape),alpha = 0.75, cmap = ListedColormap(('magenta', 'cyan')))
    ax.set_xlim(X1.min(), X1.max())
    ax.set_ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_data)):
        ax.scatter(X_data[y_data == j, 0], X_data[y_data == j, 1],
                c = np.array(ListedColormap(('white', 'black'))(i)).reshape(1,-1), label = j)
```

### Artificial Neural Networks: Brief History

#### Cybernetics 1940s-1960s

![](Perceptron1.png)
$$\text{Figure 1. Perceptron}$$

* Biologically inspired

* McCulloch-Pitts Neuron (1943)
    - Early model of brain function
    - Linear model could recognize two different categories
    - Weights set by hand
* Rosenblatt's Perceptron (1950s)
    - Learned the weights (Hebbian Learning)

* Widrow and Hoff (1960)
    - ADALINE
    - Special case of back-propagation algorithm
* Minsky and Papert (1969)
    - Proved limitations of the perceptron
        - It could not learn the XOR function
    - Killed research funding for neural compuatation
        - 1st AI Winter
#### Connectionism 1985 - early 1990's

![](MLP1.png)
$$\text{Figure 2. Multilayer Perceptron}$$


* Rumelhart and McClelland - Parallel Distributed Processing (1985)
    - Back-propagation to train multilayer perceptrons
    - Distributed representation
* Hinton (1985)
    - Boltzmann Machines -  stochastic recurrent neural network
* 2nd AI Winter
    - NN's couldn't scale beyond a couple of layers
    - Manual feature engineering
    
#### Deep Learning (2006 - present)

* Less emphasis on biological motivation
    - Don't know enough about how the brain works
* Hinton (2006)
    - Train networks with many hidden layers
        - Deep Belief Network
* Bengio (2007)
    - Training of other kinds of deep networks
* LeCunn
    - Convolutional Neural Networks
        - Learns Features

### ANN: Brief Introduction

#### Architecture

* Input layer
    - m features, ${x_i, i = 1,...,m}$
    - One input node per feature
* Hidden Layer
    - Each layer transforms via an activation function f the weighted sum of its input $a_i = f(\sum x_iw_i)$
    - function composition
    - Parameters:
        - How many hidden layers?
        - How many nodes in each layer?
* Output Layer
    - Final transformation to target outcome
    
#### Activation functions
* Logistic (i.e. Sigmoid)
* Hyperbolic tangent (tan-h)
* Rectified Linear Unit (RelU)
    - max(0,x)


```python
def ReLU(x):
    zero = np.zeros(len(x))
    return np.max([zero, x], axis=0)
x = np.arange(-2,2, .1)
plt.plot(x,ReLU(x))
plt.title("ReLU");
```

#### Learning the weights

* Learn the weights using back-propagation
    - Adjust the weights by propagating the error (loss function) backwards from the output layer
    - Compute gradients (partial derivatives of weight vectors
* Loss functions
    - The function to be minimized during training
    - Function of the output value and target value
    - Cross-entropy loss (log loss): -(ylog(p) + (1-p)(log(1-p))
        - measures the performance of a classification model that outputs a probability.
        - increases as the predicted probability diverges from the actual label.

```python
def CrossEntropy(yHat, y):
    if y == 1:
      return -log(yHat)
    else:
      return -log(1 - yHat)
```

        
#### Optimizer (i.e. solver)

* The algorithm to minimize the loss function
* Options
    - 'lbfgs': quasi-Newton method
    - 'sag': Stochastic Gradient Descent
    - 'adam': Stochastic Gradient Descent

### MultiLayerPerceptron (MLP) Logistic Regression 

http://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html


```python
# Importing the dataset
ads = pd.read_csv('knn_data.csv')
ads.head()
```

```python
X = ads.loc[:, ["Age","EstimatedSalary"]].values
y = ads.loc[:, "Purchased"].values
```

```python
# Splitting the dataset into the Training set and Test set

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.25, random_state = 1)

# Feature Scaling

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting model to the Training set

model = MLPClassifier(solver='lbfgs', activation = 'logistic' ,hidden_layer_sizes=(2), random_state=123)
model.fit(X_train, y_train) 
```

```python
print("Score: ",model.score(X_test,y_test))

# Predicting the Test set results
y_pred = model.predict(X_test)

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Plot Results
fig,(ax1,ax2)=plt.subplots(1,2,figsize = (10,6))

# Visualising the Training set results
classification_plot(X_train,y_train,model, ax1)
ax1.set_title('Training Data')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.legend()

# Visualising the Test set results
classification_plot(X_test,y_test, model, ax2)
ax2.set_title('Test Data')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary')
ax2.legend();
```

## Pytorch

```python

# Build the model with sequential
model = nn.Sequential(
    nn.Linear(2, 1),
    nn.Sigmoid()
)
```

```python
# Splitting the dataset into the Training set and Test set

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.25, random_state = 1)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Convert data into torch tensors
X_train_t = torch.from_numpy(X_train.astype(np.float32))
X_test_t = torch.from_numpy(X_test.astype(np.float32))
y_train_t = torch.from_numpy(y_train.astype(np.float32).reshape(-1, 1))
y_test_t = torch.from_numpy(y_test.astype(np.float32).reshape(-1, 1))
```

```python
torch.manual_seed(1)

# Create the classification model class, loss and optimizer

class LogRegress(nn.Module):
    def __init__(self,num_in,num_out):
        super(LogRegress, self).__init__() 
        self.linear = torch.nn.Linear(num_in, num_out)  
        self.sigmoid = nn.Sigmoid()
  
    def forward(self, x): 
        y_pred = self.linear(x) 
        return self.sigmoid(y_pred) 
    
model = LogRegress(X_train.shape[1],1)

# Loss and optimizer
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters())
```

```python
# Train the model
n_epochs = 1000

# Stuff to store
train_losses = np.zeros(n_epochs)
test_losses = np.zeros(n_epochs)

for it in range(n_epochs):
  # zero the parameter gradients
  optimizer.zero_grad()

  # Forward pass
  outputs = model(X_train_t)
  loss = criterion(outputs, y_train_t)
    
  # Backward and optimize
  loss.backward()
  optimizer.step()

  # Get test loss
  outputs_test = model(X_test_t)
  loss_test = criterion(outputs_test, y_test_t)

  # Save losses
  train_losses[it] = loss.item()
  test_losses[it] = loss_test.item()
    
  if (it + 1) % 50 == 0:
    print(f'Epoch {it+1}/{n_epochs}, Train Loss: {loss.item():.4f}, Test Loss: {loss_test.item():.4f}')
```

```python
# Plot the train loss and test loss per iteration
plt.plot(train_losses, label='train loss')
plt.plot(test_losses, label='test loss')
plt.legend()
plt.show()
```

```python
# Get accuracy
with torch.no_grad():
  p_train = model(X_train_t)
  p_train = np.round(p_train.numpy())
  train_acc = np.mean(y_train_t.numpy() == p_train)

  p_test = model(X_test_t)
  p_test = np.round(p_test.numpy())
  test_acc = np.mean(y_test_t.numpy() == p_test)
print(f"Train acc: {train_acc:.4f}, Test acc: {test_acc:.4f}")
```

```python
from matplotlib.colors import ListedColormap
def torch_plot(X_data,y_data,model,ax):
    # Make grid: min to max of 1st column, min to max of 2nd column in small increments
    X1,X2 = np.meshgrid(np.arange(X_data[:, 0].min() - 1, X_data[:, 0].max() + 1, step = 0.01),
                     np.arange(X_data[:, 1].min() - 1, X_data[:, 1].max() + 1, step = 0.01))
    # Flatten X1 and X2, create an array and transpose to a 2-column array (one for each feature)
    v = np.array([X1.ravel(), X2.ravel()]).T
    v = torch.from_numpy(v.astype(np.float32))
    v_ = model(v) #  Using fitted model, predict the points in v 
    ax.contourf(X1, X2, v_.detach().numpy().reshape(X1.shape),alpha = 0.75, cmap = ListedColormap(('magenta', 'cyan')))
    ax.set_xlim(X1.min(), X1.max())
    ax.set_ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_data)):
        ax.scatter(X_data[y_data == j, 0], X_data[y_data == j, 1],
                c = np.array(ListedColormap(('white', 'black'))(i)).reshape(1,-1), label = j)
```

```python
# Plot Results
fig,(ax1,ax2)=plt.subplots(1,2,figsize = (10,6))

# Visualising the Training set results
torch_plot(X_train,y_train,model, ax1)
ax1.set_title('Training Data')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.legend()

# Visualising the Test set results
torch_plot(X_test,y_test, model, ax2)
ax2.set_title('Test Data')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary')
ax2.legend();
```
