## Probability
 
* Quantify uncertainty  
* Probability and Statistics Inverse Relationship
    + Probability: completely specified model -> predict data from running the experiment
    + Statistics: use data from running an experiment -> draw inferrences about unknowns in model 
    
### Terminology
 
* Experiment or trial: a process with an observable outcome
* Random Experiment: An experiment whose outcome can not be predicted before the experiment is run.
* Random Variable: the distribution of the outcomes of a random experiment (process)
* Sample Space - the set of all possible outcomes
* Event: collection of possible outcomes
* Probability of an event: measure of certainty of that event
* Parameter: non-random quantity in a model
* Statistic - a measure of some attribute of a sample

### Two different interpretations of probability

#### Frequentists 
 
* Probability is a long run frequency
* Law of Large Numbers
* Grounded in real world: a sequence of events that occur in the real world
* Unambiguous
* But infinite sequences don't occur in real world
* Point estimates

####  Bayesian 
* Subjective,  degree of belief
* Degree of belief changes with more data
* Can assign probabilities to any event, doesn't have to be repeatable
* Two people can assign different probabilities to same event
* Distribution of prior probability 
* Work with entire distributions
    - contains the uncertainty


### Kolmorgorov Axioms

* Axiomatic(Formal Theory)
    + 1930's A.N. Kolmorgorov
 
* Given an event E in a sample space $S = (\cup^N_{i=1}{E_i})$
 
    + $P(E_i)$ is called the Probalbility of event $E_i$

#### Three Axioms
 
1. $0 \le{P(E_i)} \le1$
2. P(S) = 1
3. $Given: E_1 \cap{E_2} = 0\text{ then }P(E_1\cup{E_2}) = P(E_1) + P(E_2)$
    
### Properties

Denote the complement of an event A: $\bar{A} = \Omega - A$

$A \cap{\bar{A}} = \emptyset$

$1 = P(\Omega) = P(A) + P(\bar{A})\text{ that is}P(\bar{A}) = 1 - P(A)$

$P(\emptyset) = 0$

$\text{if }A\subset{B}\text{ then }P(A)\leq{P(B)}$

### Law of Additivity
 
$P(A\cup{B}) = P(A) + P(B) - P(A\cap{B}))$
 
![](additivity.png)
$$\text{Figure 1. Law of Additivity}$$



### Conditional Probability
 
P(B|A) is the conditional probability of B given A
$P(B|A) = \frac{P(A\cap{B})}{P(A)}$
 
### Chain Rule
 
$P(A\cap{B}) = P(A)P(B|A)$

$P(A\cap{B}) = P(A,B)$ is the joint probability of A and B
 


$P(B|A) = \frac{P(A\cap{B})}{P(A)}$ Denominator renormalizes the sample space
 
More generally:

$P(A,B,C) = P(A|B,C)\cdot{P(B|C)}\cdot{P(C)}$
 
### Bayes Rule

$ P(A\cap{B}) = P(B\cap{A})$
 
$P(A)P(B|A) = P(B)P(A|B)$
 
$P(B|A) = \frac{P(B)P(A|B)}{P(A)}$


### Law of Total Probability
 
$P(A) = \sum_{i}P(A\cap{B_i})$

$P(A) = \sum_{i}P(A|B_i)P(B_i)$


![](TotProb.jpeg)
$$\text{Figure 2. Law of Total Probability}$$


#### Example:

   An urn has 3 red balls and 2 black balls with equal probability of being picked. If a red ball is picked it is replaced but a black ball is not replaced. What is the probability that the second ball picked is black?

$H_1\equiv{\text{1st ball picked is red, leaving 3 red + 2 black}}$

$H_2\equiv{\text{1st ball picked is black, leaving 3 red  + 1 black}}$

$A\equiv{\text{2nd ball picked is black}}$

$P(A) = P(A|H_1)P(H_1) + P(A|H_2)P(H_2)$

$P(A) = \frac{2}{5}\cdot{\frac{3}{5}} + \frac{1}{4}\cdot{\frac{2}{5}} = 0.34$

### Independence

Events A and B are independent if
 
$P(A|B) = P(A)$
 
$P(A\cap{B}) = P(B)P(A)$

### Expected Value

* Discreet

$E(X) = \sum_x{x\cdot{p(x)}}$

* Continuous

$E(X) = \int_x{x\cdot{p(x)dx}}$


## Distributions

### Terminology
  
* Random Variable (X) - a function which maps outcomes of random processes (experiment) to a number.
    
* Probability Mass Function (PMF) - a function that calculates the probability that a discrete random variable equals a specific value P(X=x) = $f_X(x)$. Must sum to 1.
 

* Probability Density Function (PDF) - a function that calculates the relative likelihood of a continuous random variable to equal a certain value. $f_X(x)$
    + Must be positive
    + The area under the curve is the probability, it must integrate to 1
    


$$ f(x) = 3x^2 \text{ for }x\in{(0,1)}$$


$$ f(x) = x^3/4 \text{ for }x\in{(0,c)}\text{ What is c?}$$

    
* (Cumulative) Distribution Function (CDF) - a function that calculates the probability that a random variable X  will take on a value less than or equal to x. $P(X \leqq{x}) = F_X(x)$
    - For continuous probability distributions it is the integral of the pdf.
$$  P(2\leqq{X}\leqq{4}) = \int^4_2f(x)dx$$

### Binomial Distribution 
 
* If I flip a fair coin twenty times, what is the probability of 7 heads?
    - Binomial(N,p)
* The number of successes in N tries.
* The PMF is:
 
$P(X=k | N;p) = \frac{N!}{k!(N - k)!}p^k(1 - p)^{(N - k)}$
 
* X - random variable
* k - # successes(7) 
* N - size parameter (20) 
* p - success probability
 
#### Moments: 
 
* Mean = np
* Variance = np(1-p)=npq 
    - Note: (q = 1 - p)


#### Binomial in R

```python
# Plot PMF
print("P(X=7):")
prob = dbinom(x = 7, size = 20, prob = 1/2);prob
x <- seq(0,20,by=1)
y <- dbinom(x,20,0.5)
plot(x,y,pch=16,type='h',main = 'PMF for Binom(20,.5)',ylab="Probability", xlab='Size')
points(7,prob,pch=16,col='red')
```

```python
p = pbinom(7,20,0.5)
q = qbinom(p,20,0.5)
c("p"=p,"q"=q)
x= seq(0:20)
plot(x,pbinom(x,20,0.5),pch = 16, type = 'b',
     ylab="Probability", xlab='Size',main = 'CDF for Binom(20,.5)')
lines(c(7,7),c(0,p))
lines(c(0,7),c(p,p))
points(q,p,pch=16,col='red')
```

```python
r = rbinom(100,size=20,p=0.5)
hist(r, main = 'Histogram for Binom(20,.5)',xlab="Size",ylab='Frequency')
```

```python
set.seed(1234)
flip = function(p=.5) {sample(1:100,1) > 100*p}
n.flips = function(n,p) {sum(replicate(n,flip(p)))}
sim.pmf.binom = function(k,n,p = .5, N_Iters=10000) {
  total = replicate(N_Iters,n.flips(n,p))
  sum(total == k)/N_Iters
}
sim.pmf.binom(7,20)
```

```python
import numpy as np

np.random.seed(1234)

def flip (p = 0.5):
    return (np.random.choice(100) >= 100*p)

def nFlips (n, p = 0.5):
    return sum(np.array([flip(p) for i in range(n)]))

def simPmfBinom (k,n,p=0.5,N_Iters = 10000):
    total = np.array([nFlips(n,p) for i in range(N_Iters)])
    return( sum(total == k)/N_Iters)

simPmfBinom(7,20)
```

### Bernouilli Distribution
 
* Bernoulli Trial - a random experiment with exactly two possible outcomes, {success, failure}
* Bernoulli is Binom(1,p)
* Distribution of a RV that take a value of 1 with probability p and a value of zero with probability (1-p)
* The PMF is:
 
$ pmf(k;p) = p^k\cdot{(1 - p)^{(1-k)}\text{ for k}\in{\{0,1\}}}$
* Mean and Variance
    - E(X) = p 
    - Var(X) = p(1-p)  
* If $X_1,X_2,...,X_n$ are independent, indentically distributed (i.i.d.) Bernoulli random variables with
success probability p, then there sum is a Binomial random variable Y, such that

 
$Y = \sum^n_{k=1}X_k\sim{Binom(n,p)}$
    


### Negative Binomial

* Number of trials k needed to obtain r successes in repeated independent Bernoulli trials
    
$P(X=k) = \frac{(k-1)!}{(r-1)!(k-r)!}p^r(1-p)^{k-r}\text{  for k = r, r+1, r+2,...}$

$Mean = \frac{p*r}{1-p}$
 
$Variance = \frac{p*r}{(1-p)^2}$

* Flip a fair coin repeatedly until have 5 heads, what is the probability that the 5th head occurs on the 7th trial?

```python
# R code
dnbinom(7,5,.5)
v <- dnbinom(5:20,5,.5)
names(v) <- 5:20
barplot(v, xlab="Trial", ylab="Probaility", main= "Negative Binomial(7,.5)")
```

```python
# Python Code
from scipy.stats import nbinom
import matplotlib.pyplot as plt
import numpy as np
k,n,p = 7,5,.5
print(nbinom.pmf(k,n,p))
x = np.arange(5,21,step=1,dtype='int')
v = nbinom.pmf(x,n,p)
plt.bar(x,v)
plt.title("Negative Binomial(7,.5)")
plt.xlabel("Trial")
plt.ylabel("Probability")
plt.show()
```

###  Geometric

* Negative Binomial with r = 1
 
#### Two definitions
 
* Probability distribution of number (k = 0,1,2,...) of failures before 1st success
    - R uses this definition

$pmf: P(X=k) = p*(1 - p)^k \text{ for k = 0,1,2,...}$

* Probability distribution of trial number (k = 1,2,...) of 1st sucesss
    - Python uses this definition
 
$pmf: P(X=k) = p*(1 - p)^{(k-1)}\text{ for k = 1,2,...}$


$Mean = \frac{(1-p)}{p}$

$ Variance = \frac{(1 - p)}{p^2}$
 
* Flip a biased coin (p = .2). Heads (x = 1) is a success.
    + What is probabilty that the first six flips will be tails?
    + What is the expected number of flips before a tail?
    + What is the probability that a head will occur in the first 6 flips?


```python
# R code
k = 6
p = .2
print("Probability first 6 flips will be tails:")
dgeom(k,p) 
q = 1 - p 
print("Expected number of flips before a tail:")
q/p #mean
print("Probability that a head will occur in the first 6 flips:")
pgeom(k-1,p)
# Just sum pmf to get cdf
sum(sapply(0:(k-1),dgeom,p)) #Note geom takes two arguments
```

```python
# Python code
from scipy.stats import geom
import numpy as np
# Python uses 2nd definition (k - 1) so add 1 to number of flips
k,p,q = 7,0.2,(1-0.2)
pmf = geom.pmf(k,p)
print("Probability first 6 flips will be tails:",pmf)
mean = q/p
print("Expected number of flips before a tail:",mean)
cdf = geom.cdf(k-1,p)
print("Probability that a head will occur in the first 5 flips:",cdf)
x = np.arange(0,k,step=1,dtype='int')
sum(i for i in map(lambda k : geom.pmf(k,p),x))


```

## Poisson Distribution

* Probability of a given number or events in a fixed interval of time(and/or space). Events occur at a known average rate and independently of the time since the last event.

* Examples
    - number of calls received by a call center per hour
    - number of people in queue for a teller at a bank


$\lambda$ is the shape parameter which indicates the average number of events in the given time interval

$$P(X=x) = \frac{\lambda^x\cdot{e^{-\lambda}}}{x!}$$

Nice Property $\lambda = E(X) = Var(X)$

The Apple store sells ten iphones per day. What is the probability that exactly 6 iphones will be sold tommorrow


```python
# R code
print("Probability that exactly 6 iphones will be sold tommorrow:")
dpois(x = 6, lambda = 10)
x=seq(0,20,1)
p = dpois(x,10)
plot(x,p,pch=16,type='h', main="Poisson(10)",xlab="Number of iphones",ylab="Probability")
```

```python
# Python code
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 10 # Can not use lambda
k = 6

pmf = poisson.pmf(k,mu)
print("Probability that exactly 6 iphones will be sold tommorrow:",pmf)

x = np.arange(0,21,step=1,dtype='int')
plt.plot(x, poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
plt.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
plt.show()

```

### Normal (Gaussian) Distribution
 
* Most common continuous distribution
    - A process that repeatedly adds together fluctuations will tend towards a Normal
        - Any process that adds together random values from the same distribution converges to a normal. 
    - related to real-valued quanities that grow linearly
        - errors
    - Central Limit Theorem
        - if you take repeated samples from a population with finite variance and calculate their averages, then the averages will be normally distributed
        
$$X \sim N(mean=\mu,sd = \sigma)$$    
    
#### Probability Density Function (PDF): 
 
* Integrates to 1

$$pdf:  f(x | \mu, \sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

```python
# R code
curve(dnorm(x),-3,3,main="Normal(0,1) PDF",ylab="Density")
```

```python
# Python code
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,100)
y = norm.pdf(x,loc=0,scale=1)
plt.plot(x,y)
plt.title("Normal(0,1)")
plt.xlabel("X")
plt.ylabel("Density")
plt.show()
```

###  Central Limit Theorem
 
*  The distribution mean of a large number of i.i.d. samples from any distribution will converge to a Normal distribution as the sample size -> infinity
    + e.g. if i repeatedly sample from a geometric distribution, the means of those samples will be normally distributed


```python
# R code
# Change sample size from 2,10, 50
set.seed(1234)
means2 = sapply(1:10000,function(i) mean(rgeom(2,prob=.5)*10))
means10 = sapply(1:10000,function(i) mean(rgeom(10,prob=.5)*10))
means50 = sapply(1:10000,function(i) mean(rgeom(50,prob=.5)*10))
means100 = sapply(1:10000,function(i) mean(rgeom(100,prob=.5)*10))
par(mfrow=c(2,2))
hist(means2,main= "n=2")
hist(means10,main="n=10")
hist(means50,main="n=50") # Already looks pretty normal
hist(means100,main="n=100")

```

```python
# Python code
from scipy.stats import geom
import matplotlib.pyplot as plt
import numpy as np

fig =  plt.figure()

def mapMeanGeom(sample_size,p = .5): 
    return(np.mean(geom.rvs(p,size=sample_size)))
    

means2 = [mapMeanGeom(2) for i in range(1000)]
means10 = [mapMeanGeom(10) for i in range(1000)]
means50 = [mapMeanGeom(50) for i in range(1000)]
means100 = [mapMeanGeom(100) for i in range(1000)]

ax1 = fig.add_subplot(2,2,1)
ax1.hist(means2)
plt.title("Means2")
ax1 = fig.add_subplot(2,2,2)
ax1.hist(means10)
plt.title("Means10")
ax1 = fig.add_subplot(2,2,3)
ax1.hist(means50)
plt.title("Means50")
ax1 = fig.add_subplot(2,2,4)
ax1.hist(means100)
plt.title("Means100")
plt.show()

```

#### Stanard Normal

* $X \sim N(0,1)$

* z-scores; number of standard deviations from mean
* Convert y from distribution with mean $\hat{y}$ and standard deviation s: $z = \frac{y - \hat{y}}{s}$

Measure the heights of 100 people, mean = 170 cm, sd = 8 cm.

What is probability that a random selected person will be less than 160 cm in height?

$z = \frac{160 - 170}{8} = -1.25$

### Uniform distribution
 
* Equal probability in an interval
    - Defaults to[0,1]
    
#### Probability Density Function

$$f(x) = \frac{1}{b-a}\text{ for x}\in [a,b]\text{ 0 otherwise}$$
* Mean: $\frac{1}{2}\cdot(a+b)$  
* Variance: $\frac{1}{12}\cdot{(b-a)^2}$

```python
# R code
set.seed(1234)
flip = function(p=.5) {return(if (runif(1) <  p) 1 else 0)}
print("Flipping a .8 biased coin 1000 times:")
sum(replicate(1000,flip(.8))/1000)
x <- runif(100000)
y <- runif(100000)
plot(x,y,pch=".",col="blue", main="2-D Uniform(0,1)")
```

```python
# Python code
np.random.seed(1234)
from scipy.stats import uniform
import matplotlib.pyplot as plt

def flip(p=0.5):
    return(1 if uniform.rvs(loc=0,scale=1,size=1) < p else 0)
flips = np.array([flip(.8) for i in range(1000)])
print("Flipping a .8 biased coin 1000 times: %3f" % float(sum(flips)/1000))

x = uniform.rvs(size=1000)

y = uniform.rvs(size=1000)
plt.plot(x,y,color='b',marker = '.')
plt.title("2-D Uniform(0,1)")
plt.show()
```

###  Universality of the Uniform (Probability Integral Transform)
 
* When you plug a **continuous** r.v. into its own CDF you get a Uniform
    - F(X) ~ U
* When you plug a Uniform r.v. into the inverse CDF of X you get a r.v. distruted like X
    - $F^{-1}(U) \sim X$
    * Only works for distributions whose CDFs are continuous and strictly-increasing
  

```python
# R code
# F(X) ~ U
d = rnorm(1000)
u <- pnorm(d)
print("X ~ norm(0,1), F is CDF of norm(0,1)")
par(mfrow=c(1,2))
hist(u, main = 'F(X) ~ U')
# F_inverse(U) ~ X
dd <- runif(1000)
ddd <- qnorm(dd,200,.5)
print("U ~ Uniform(0,1), F-inverse is normal quantile function")
hist(ddd,main = 'F-inverse(U) ~ X',xlab = "Normal R.V.",ylab = 'Frequency')
```

```python
# Python code
from scipy.stats import norm, uniform
import matplotlib.pyplot as plt
import numpy as np

fig =  plt.figure()

# F(X) ~ U
d = norm.rvs(200,.5, size = 1000)
u = norm.cdf(d,200,.5)
print("X ~ norm(0,1), F is CDF of norm(0,1)")
ax1 = fig.add_subplot(1,2,1)
ax1.hist(u)
plt.title('F(X) ~ U')
# F_inverse(U) ~ X
dd = uniform.rvs(size = 1000)
ddd = norm.ppf(dd,200,.5)
print("U ~ Uniform(0,1), F-inverse is the normal quantile function")
ax2 = fig.add_subplot(1,2,2)
ax2.hist(ddd)
plt.title('F-inverse(U) ~ X')
plt.xlabel("Normal R.V.")
plt.ylabel('Frequency')
plt.show()
```

### Beta Distribution
 
#### A probability distribution of probabilities
 
* It represents all the possible values of a probability when we don’t know what that probability is
    - a real number between 0 and 1
* Flexible
    - Different structures as a function of shape paraeters a and b.
        - Beata(1,1) is Unform(0,1)
 
#### Baseball batting averages
* The number of times a player gets a base hit divided by the number of times he goes up at bat
* A percentage between 0 and 1. 
* .266 is average, > .300 is excellent
* Binomial distribution (success = gets a hit)
* Predict what season long batting average will be
    - at beginning 1 hit in ten at bats or 6 hits in 10 at bats will not be very predictive
        - have some knowledge about range of batting averages (.210 to 350)
* Can use beta distribution to reflect this prior knowledge
 
#### X ~ Beta(a,b) a - number of successes, b - number of failures
 
$$ PDF = f(x;a,b) = \frac{1}{B(a,b)}x^{a-1}*(1-x)^{b-1} \text{ for a> 0, b>0, x } \in{[0,1]}$$
 $$ BetaFunction=B(a,b) = \frac{\Gamma(a)\Gamma(b)}{\Gamma(a+b)}$$
 
* Mean: $\frac{a}{a+b}$
* Variance: $\frac{a*b}{(a + b)^2*(a + b + 1)}$
  


```python
# R code
par(mfrow=c(2,2))
curve(dbeta(x,1,1)) # Uniform
curve(dbeta(x,3,2))
curve(dbeta(x,1/2,1/2))
curve(dbeta(x,2,2))
```



##### Properties
 
* $\Gamma$(n) = (n-1)! for positive integer n
* $\Gamma$(x+1) = x$\Gamma$(x)
* $\Gamma(1/2) = \sqrt{\pi}$


```python
# R code
n = 6
c(gamma(n),factorial(n-1))
x = pi
c(gamma(x+1),x*gamma(x))
c(gamma(1/2),sqrt(pi))
```

### Gamma Distribution

$$ 1 = \int_0^\infty{\frac{1}{\Gamma(a)}x^{a-1}e^{-x}dx}$$
$$ PDF = Gamma(a,1) = \frac{1}{\Gamma(a)}x^{a-1}e^{-x}$$
$$ PDF = Gamma(a,b) = f(x;a,b) = \frac{1}{\Gamma(a)}b^ax^{a-1}e^{-b x}\text{ for x > 0 } $$

#### Two parameters (a,b)
 
* a: shape parameter
* b: scale parameter (1 / rate) (note: rate = $\lambda$)

```python
# R code
par(mfrow=c(2,2))
curve(dgamma(x,1,1),0, 10, main = "Gamma(1,1)",ylab = "Density")
curve(dgamma(x,2,1),0,10, main = "Gamma(2,1)",ylab = "Density")
curve(dgamma(x,2,2),0,10, main = "Gamma(2,2)",ylab = "Density")
```

```python
# Python code
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np

fig =  plt.figure()
ax1 = fig.add_subplot(2,2,1)
x = np.linspace(0,10,100)
ax1.plot(x,gamma.pdf(x,1,loc=0, scale=  1))
ax2 = fig.add_subplot(2,2,2)
ax2.plot(x,gamma.pdf(x,2,loc = 0,scale = 1))
ax3 = fig.add_subplot(2,2,3)
ax3.plot(x,gamma.pdf(x,2, loc= 0,scale =1/2))
plt.show()
```

####  Relationship to Poisson Process

* If X ~ Pois($\lambda$) then time until k arrivals is Gamma(k,$\frac{1}{\lambda}$)
    
 
##### Example
 
Suppose customers arrive at a store as a Poisson process with rate 10 customers per hour.
What the probability that one has to wait at least half an hour until the second customer arrives?
    

```python
# R code
lambda = 10 # 10 customers per hour
k = 2 # Second customer
curve(dgamma(x,k,scale=1/lambda),0,1,ylab='Density',main='Gamma(2,1/10)')
x = 1/2
# P(T >= 1/2) = 1 - P(T < 1/2)
1 - pgamma(x,k,scale=1/lambda)

y = dgamma(x,k,scale=1/lambda)
points(x,y,col = 'red',pch = 16)
x = seq(.5,1.0,.01)
y = dgamma(x,k,scale=1/lambda)
polygon(c(x, .5 ), c(y,0), col="red")
```

```python
# Python code
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np
mu = 10 # 10 customers per hour
k = 2 # Second customer
x= np.linspace(0,1,100)
plt.plot(x,gamma.pdf(x,k,scale = 1/mu))
plt.ylabel='Density'
plt.title('Gamma(2,1/10)')
x = 1/2
# P(T >= 1/2) = 1 - P(T < 1/2)
1 - gamma.cdf(x,k,scale = 1/mu)

y = gamma.pdf(x,k,scale = 1/mu)
plt.plot(x,y,'bo', color = 'red')
x = np.linspace(.5,1.0,50)
y = gamma.pdf(x,k,scale = 1/mu)
plt.fill_between(x,y,color="r")
plt.show()
```
