## What is Statistical(Machine) Learning?
 
* $Y = f(X) +\epsilon$
    - Y: response(dependent) variable
    - X: predictor(independent) variable  $X = (X_1,X_2,...,X_p)$  
    - $\epsilon$: random error term, independent of X and mean of 0  
        + unmeasurable variables, unmeasurable variation
    - f: fixed but unknown function of X 
        - the systematic info that X provides about Y

* Statistical learning is a set of approaches for estimating f in order to predict Y and/or make inferences

### Model Complexity

* Simpler models are easier to interpret
    - Occam's razor:  if we have two or more equivalent explanations for the same phenomenon, we should choose the simpler one.
    - But can underfit the data

* Models with more parameters (degrees of freedom) are more flexible but are less interpretable (not good if inference is the goal)
* If prediction is the goal, more flexible models tend to be more accurate but not always (overfitting)
    
* The more parameters a model has, the more complex it is since it can make more predictions
* Models that are more complex are more flexible, they can fit different datasets (i.e. more accurate, higher $R^2$)
    - But are prone to overfitting
* Overly complex models are able to make many predictions, but many of them are false
* We want a model that is high on accuracy but is as simple as possible
    - "Things should be as simple as possible but not simpler" Albert Einstein

#### Example

```python
plt.style.use(['seaborn-darkgrid'])
x_n = np.linspace(4,14, 100)
order = [0, 1, 2, 5]

x = np.array([4.,5.,6., 9.,12, 14.])
y = np.array([4.2, 6., 6., 9., 10, 10.])

def plot_poly(x,y):
    plt.figure(figsize=(7, 7))
    plt.plot(x, y, 'o')
    for i in order:
        coeffs = np.polyfit(x, y, deg=i)
        ffit = np.polyval(coeffs, x_n)
        p = np.poly1d(coeffs)
        yhat = p(x)
        ybar = np.mean(y)
        ssreg = np.sum((yhat-ybar)**2)
        sstot = np.sum((y - ybar)**2) 
        r2 = ssreg / sstot
        plt.plot(x_n, ffit, label='order {}, $R^2$= {:.2f}'.format(i, r2))
    plt.legend(loc=2, fontsize=14)
    plt.xlabel('$x$', fontsize=14)
    plt.ylabel('$y$', fontsize=14, rotation=0)

plot_poly(x,y)

```

* The model increases in complexity as more terms (parameters) are added and $R^2$ increases explaining more of the variance in y

#### Overfitting
 
* Model with order 5 fits the data perfectly
    - Same number of parameters as data points
    - Notice the erratic behavior of this model between the data points (9,9) and (12,9).
* The behavior of order 2 model is much smoother, it predicts a parabola with some noise
    - This seems a more reasonable model even though it has a lower $R^2$
* If we collected more data points, say (10,9) and (7,7)

```python
x = np.array([4.,5.,6., 7, 9.,10,12, 14.])
y = np.array([4.2, 6., 6., 7, 9., 9, 10, 10.])
plot_poly(x,y)
```

* Notice how the order 5 model changed shape and its $R^2$ went down.
* Order 1 and 2 models changed very little. They were better predictors of the new data
* Models that overfit the data have high variance.

#### Underfitting

* Model with order 0 has no coefficients so the model is just a model of the dependent variable alone. The green line represents the mean of the dependent variable y. It is not capturing any patterns in the data.
* Models that underfit the data, like the order 0 and order 1 models, have high bias



### Model Accuracy

#### Prediction

* $\hat{Y} = \hat{f}(X)$ is some model that estimates f(x)
* Accuracy of $\hat{f}(x)$ in predicting Y
    - reducible error ($\hat{f}$ is not a perfect estimator)
    - irreducible error ($\epsilon$ which is independent of X)
    
<div style="font-size: 115%;">
$$E(Y - \hat{Y})^2 = E[f(X) + \epsilon - \hat{f}(X)]^2  =  [f(X) - \hat{f}(X)]^2 + var(\epsilon)$$
</div>

* $E(Y - \hat{Y})^2$ represents the average squared difference of actual and estimated value (reducible error) and the variance of the error term assuming mean = 0 for distribution of errors (irreducible error)
* The irreducible error can't be minimized, so goal is to minimize the reducible error 
    
#### Mean Squared Error (MSE): measures the quality of fit  

<div style="font-size: 115%;">
$$MSE = \frac{1}{n}\sum^n_{i=1}(y_i-\hat{f}(x_i))^2$$ 
</div>

* Calculated on the training data
* But what we care about is how well the method does on the test data (i.e. previously unseen data)
    - Does the model generalize
* No guarantee that is training MSE is small that the test MSE is small

![](MSE.png)
$$\text{Figure 3. Within-sample (gray curve) verses out-of-sample (red curve)}$$
 
* In the figure, on the left are three different models of varying complexity
* On the right: orange line is MSE on test data, blue line is MSE on the training data
    - For training data, the MSE will always decrease as the complexity of the model increases
    - For test data, a U-shaped curve

#### Within-sample versus out-of-sample accuracy

* **Within-sample** accuracy is the accuracy of model fit to the data. It will always increase with more parameters
* **Out-of-sample** accuracy is the accuracy of the model on new data. When it starts to decrease (error starts to increase) is the point at which it is overfitting the data.


#### Overfitting
 
* Overfitting: When the learning method is finding patterns in the training data caused by random chance (noise) rather than properties of the unknown function f  
    - U-shaped curve  
    - Small training MSE and large test MSE indicates overfitting  
    - Cross validation - method for estimating test error using the training data 
* Overfitting occurs when **out-of-sample** mean squared error (MSE) increases
    - MSE = squared difference of actual data value and value predicted by the model averaged over all data points 

 

## Bias-Variance tradeoff

$$Y = f(x) + \epsilon \text{ where } \epsilon \sim N(0,\sigma)$$

* The outcome Y is a function of some independent variable x and some error $\epsilon$
* f(x) is a real-valued function so this is a regression setting
* $\hat{f}(x)$ estimates f(x) (i.e. is a model for f(x))
* Given K datasets $D_1,D_2,...,D_K$ where a dataset is a sample of the population
* Let $\bar{f}(x) = E_D[\hat{f}_D(x)$ i.e. the average over all the datasets of $\hat{f}_D(x)$ 
    - $ \bar{f}(x) \approx \frac{1}{K}\sum_{k=1}^K\hat{f}_{D_k}(x)$
* Expected Test MSE

<div style="font-size: 115%;">
$$Err(x) = E_D[(\hat{f}_D(x) - Y)^2] $$  


$$ = E_D[(\hat{f}_D(x) - \bar{f}(x))^2] + (\bar{f}(x) - f(x))^2 +  \sigma^2$$  

$$ = Variance + Bias^2 + Irreducible Error$$  

</div>

* The expected Test Error can be decomposed into
    - Variance: the average squared difference of the model on a particular dataset D and the mean of the model on K datasets
    - Bias: The difference of the models average estimate from the true value
    - Irreducible Error: The error that can't be reduced even if we had a better model.

![](bias-and-variance.jpg)
$$\text{Figure 4. Visualize Bias and Variance}$$
 
![](model-complexity.png)
$$\text{Figure 5. Bias and Variance Curves}$$

* To minimize test error need low variance and low bias  
* Variance: the amount $\hat{f}$ would change if used different training set. 
    - More flexible (i.e. more complex, more paramters) methods have higher variance
        - Tuned to the sample, if get a different sample, will get different reducible error
* Bias: the error introduced by modeling a complicated process with a simpler model
    - Linear regression assumes a linear relationship but most real world problems are not purely linear
    - More flexible models have less bias
    - As we increase flexibility bias tends to decrease faster than variance increases (lower test MSE) but at some point, increasing flexibility has little effect on bias but significantly increases variance.
